```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import os
import numpy as np

saves = os.listdir('.')
saves.sort(key=os.path.getmtime)
saves.reverse()
saves = [save for save in saves if os.path.isdir(save)]
```

```python
# reward_log = saves[0] + '/rewards.log'
reward_log = 'simpleDQN' + '/rewards.log'
with open(reward_log, 'r') as f:
    scores = [float(score.strip()) for score in f.readlines()]
sns.regplot(np.array(range(len(scores))), np.array(scores))
# sns.regplot(np.array(range(len(scores[:80]))), np.array(scores[:80]))
```
