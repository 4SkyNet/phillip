```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import os
import numpy as np
from IPython.display import display, HTML

import matplotlib.pylab as pylab
pylab.rcParams['figure.figsize'] = 15, 9

def header(text, level, inline=True):
    tag = 'h' + str(level)
    if inline:
        return '<' + tag + ' style="display:inline-block">' + text + '&nbsp;&nbsp;</' + tag + '>'
    else:
        return '<' + tag + '>' + text + '</' + tag + '>'
```

```python
def show_results(reward_log):
    with open(reward_log, 'r') as f:
        data = [list(map(float, score.split(' '))) for score in f.readlines()]
    score_array = np.array(data)

    x_array = score_array[:, 0]
    x_array -= np.min(x_array)
    y_array = score_array[:, 1]
    sns.regplot(x_array, y_array, fit_reg=False)
    
    tail = len(score_array)//4
    x_array = x_array[-tail:]
    y_array = y_array[-tail:]
    sns.regplot(x_array, y_array, scatter=False)

    coefficients = np.polyfit(x_array, y_array, 1)
    slope = coefficients[0]
    display(HTML(header(reward_log, 1, False)))
    display(HTML(header('Score:', 2) + header(str(np.mean(y_array)), 2) +
                 header('Slope:', 2) + header(str(slope), 2)))
```
